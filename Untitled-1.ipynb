```python
import threading
import time
def number():
    for i in range(5):
        print(i)
        time.sleep(1)
def alphabet():
    for ch in ["A", "B", "C", "D", "E"]:
        print(ch)
        time.sleep(0)
t1 = threading.Thread(target=number)
t2 = threading.Thread(target=alphabet)
t1.start()
t2.start()
t1.join()
t2.join()   
print("finished")
```
